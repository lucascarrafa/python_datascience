# Dados perdidos

```python
import numpy as np
```

```python
dados = {
    'nome': ['João', 'Maria', 'José', np.nan, 'Pedro', 'Judas', 'Tiago'],
    'sexo': ['M', 'F', 'M', np.nan, 'M', 'M', np.nan],
    'idade': [14, 13, np.nan, np.nan, 15, 13, 14],
    'nota': [4, 10, 7, np.nan, 8, 9, 7]
}
df = pd.DataFrame(dados)
```

```python
df
```

```python
df.dropna() #exclui dados faltantes
```

```python
df.dropna(how='all') #apenas linhas com nenhum dado
```

```python
df['series'] = np.nan
```

```python
df
```

```python
df.dropna(how='all',axis=1) #remove coluna
```

```python
df.dropna(thresh=3) #mais de três dados faltando
```

```python
df['series'].fillna(8,inplace=True)
```

```python
df
```
